```python
data = pd.read_csv('final.csv', sep=',',names=['id', 'date','user','text'])
data.drop(data.index[:1],inplace=True)
```

```python
data
```

```python
from nltk import word_tokenize
from string import punctuation

## Transfrom all words into lowercase 
def preprocess(x):
    if type(x)!=float:
        
            
        s = x.lower()
        return s
    else:
        return x

def toeknizeds(s):
    if type(s)==str:
        s=word_tokenize(s)
        return s
        
token = data['text'].map(preprocess).map(toeknizeds)
token
```

```python
from nltk.corpus import stopwords 

stop_words = set(stopwords.words('english'))

filtered_sentence = [] 

```

```python
#drop the punctuation and stopwords
def remove_stopwords(x):
    if x!=None:
        for w in x:
            if w not in punctuation:
 
                if w not in stop_words: 
                    filtered_sentence.append(w) 
  
        
            

token.map(remove_stopwords)
filtered_sentence
```

```python
from collections import Counter
#counts the number of mentions each brand has and displays the top 10

counts = Counter(filtered_sentence)
counts.most_common
```
