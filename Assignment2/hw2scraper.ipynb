```python
import os
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import numpy as np
import pandas as pd
```

```python
def get_date(soup):
    date = []
    for element in soup.find_all(class_='Comment'):
        date.append(element.find_all('div')[3].text.strip())
    return date
```

```python
def get_userid(soup):
    userid = []
    for element in soup.find_all(class_='Comment'):
        userid.append(element.find_all('span')[0].text.strip())
    return userid
```

```python
def get_message(soup):
    msg = []
    for element in soup.find_all(class_='Comment'):
        msg.append(element.find_all('div')[6].text.strip())
    return msg
```

```python
def get_data (url, driver, pages):
    date = []
    userid = []
    msg = []
    for i in range (pages):
        driver.get(url)
        soup = BeautifulSoup(driver.page_source,'html.parser')
        date += get_date(soup)
        userid += get_userid(soup)
        msg += get_message(soup)
        next_button = soup.find("a", class_="Next Pager-nav")
        url = next_button['href']
    return pd.DataFrame(list(zip(date, userid, msg)))
```

```python
url = "https://forums.edmunds.com/discussion/2864/general/x/entry-level-luxury-performance-sedans"
driver = webdriver.Chrome(chrome_path)
get_data(url, driver, 10)
```
